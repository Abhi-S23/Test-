```python
final=[]
for index,rows in df.iterrows():
    rest_name=rows['restaurant_name']
    loc=rows['location']
    menu=rows['menu']
    for i in menu :
        final.append([rest_name , loc , i['food_name'] ,i['ingredients'] , i['price'] ])
df=pd.DataFrame(data=final,columns=['restaurant_name','location','food_name','ingredients','price'])
df
```

```python
dicti={}
for index,rows in df.iterrows():
    llist=rows['ingredients']
    for i in llist:
          
        if i in dicti:
            dicti[i]+=1
        else:
            dicti[i]=1
            

```
